```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение данных из текстового файла
file_path = './Новый текстовый документ.txt'
data = pd.read_csv(file_path, sep=" ", header=None, names=[
    "Size",
    "MergeSort_Random", "HybridSort_Random",
    "MergeSort_Reversed", "HybridSort_Reversed",
    "MergeSort_NearlySorted", "HybridSort_NearlySorted"
])

# Построение графиков
categories = ["Random", "Reversed", "NearlySorted"]
for category in categories:
    plt.figure(figsize=(10, 6))
    plt.plot(data["Size"], data[f"MergeSort_{category}"], label=f"MergeSort ({category})")
    plt.plot(data["Size"], data[f"HybridSort_{category}"], label=f"HybridSort ({category})")
    plt.xlabel("Array Size")
    plt.ylabel("Time (microseconds)")
    plt.title(f"Performance Comparison: MergeSort vs HybridSort ({category})")
    plt.legend()
    plt.grid(True)
    plt.show()
    plt.savefig(f"Performance Comparison: MergeSort vs HybridSort ({category})")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение данных из файла
file_path = './Новый текстовый документ (2).txt'
data = pd.read_csv(file_path, sep=" ", header=None, names=[
    "Size", "Threshold", "Random", "Reversed", "NearlySorted"
])

# Уникальные значения порогов
thresholds = data["Threshold"].unique()

# Построение графиков для каждого порога
for threshold in thresholds:
    subset = data[data["Threshold"] == threshold]
    plt.figure(figsize=(10, 6))
    plt.plot(subset["Size"], subset["Random"], label="Random Array")
    plt.plot(subset["Size"], subset["Reversed"], label="Reversed Array")
    plt.plot(subset["Size"], subset["NearlySorted"], label="Nearly Sorted Array")
    plt.xlabel("Array Size")
    plt.ylabel("Time (microseconds)")
    plt.title(f"Hybrid Merge Sort Performance (Threshold = {threshold})")
    plt.legend()
    plt.grid(True)
    plt.show()
```
